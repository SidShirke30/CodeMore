```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Data cleaning

```python
##check null values/missing values
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## remov the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(columns='index')
```

```python
df.head()
```

```python
df.columns
```

```python
## fix spaces in column name
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

## change the required columns as int datatype

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### changing the other columns to float datatype

```python
 objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i]=df[i].astype(float)
```

```python
##lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv',index=False)
```

## Exploratory Data Analysis

```python
## drop day , month and year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
## categoried in classes
df_copy['Classes'].value_counts()
```

```python
## encoding of the categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
##plot density plot for all features
plt.style.use("seaborn-v0_8")
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## percentage of piechart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
##plot pie chart
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie chart of classes')
plt.show()
```

```python
##correlation
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
plt.show()
```

```python
##box plot
sns.boxplot(df_copy['FWI'],color='blue')
plt.show()
```
